```python
def Delta(r, GM, a): return r**2 - 2*GM*r + a**2*GM**2

def rplus(GM, a): return GM*(1. + sqrt(1 - a**2))

def rmins(GM, a): return GM*(1. - sqrt(1 - a**2))

def rstar(r, GM, a):

    if a == 0.:
        return r + 2.*GM*log(r/(2.*GM) - 1.)
    else:
        return (r + (rplus(GM,a)**2 + a**2*GM**2)/(rplus(GM,a) - rmins(GM,a))*log(r - rplus(GM,a))
                  - (rmins(GM,a)**2 + a**2*GM**2)/(rplus(GM,a) - rmins(GM,a))*log(r - rmins(GM,a)))
    
def Omega(GM, a): return a/(2.*GM*(1. + sqrt(1 - a**2)))


def TBH(GM, a): 
    
    kappa_inv = 2.*GM*(1. + 1./sqrt(1 - a**2))
    
    return 1./(2*pi*kappa_inv)
```

### Finding Angular eigenvalues/separation constants ###

```python
def h(l,m,s):
    return (l**2 - np.max([abs(m),abs(s)])**2)*(l**2-(m*s/np.max([abs(m),abs(s)]))**2)*(l**2-s**2)/(2*(l-0.5)*(l+0.5)*l**3)

def taylor(l,m,s,c):

    f0 = l*(l+1) - s*(s+1)

    f1 = -2*m*s**2/(l*(l+1)) if l!=0 else 0

    f2 = -1 if m==0 and s==0 else h(l+1,m,s) - 1 if l==0 else h(l+1,m,s) - h(l,m,s) - 1

    f3 = 0 if l==0 or (m==0 and s==0) else -(2*h(l+1,m,s)*m*s**2)/(l*(l+1)**2*(l+2)) if l==1 else (2*h(l,m,s)*m*s**2)/((l-1)*(l+1)*l**2) - (2*h(l+1,m,s)*m*s**2)/(l*(l+1)**2*(l+2))

    f4 = 0 if (m==0 and s==0) else -(l+2)*h(l+1,m,s)*h(l+2,m,s)/(2*(l+1)*(2*l+3)) + h(l+1,m,s)**2/(2*(l+1)) if l==0 else m**2*s**4*4*h(l+1,m,s)/((l**2)*(l+1)**4*(l+2)**2) - (l+2)*h(l+1,m,s)*h(l+2,m,s)/(2*(l+1)*(2*l+3)) + h(l+1,m,s)**2/(2*(l+1))+h(l,m,s)*h(l+1,m,s)/(2*l**2+2*l) - h(l,m,s)**2/(2*l) if l==1 else m**2*s**4*(4*h(l+1,m,s)/((l**2)*(l+1)**4*(l+2)**2) - 4*h(l,m,s)/((l-1)**2*l**4*(l+1)**2)) - (l+2)*h(l+1,m,s)*h(l+2,m,s)/(2*(l+1)*(2*l+3)) + h(l+1,m,s)**2/(2*(l+1))+h(l,m,s)*h(l+1,m,s)/(2*l**2+2*l) - h(l,m,s)**2/(2*l) + (l-1)*h(l-1,m,s)*h(l,m,s)/(4*l**2-2*l)

    f5 = m**3*s**6*(-8*h(l+1,m,s)/(l**3*(l+1)**6*(l+2)**3)) + m*s**2*((3*l+7)*(h(l+1,m,s)*h(l+2,m,s))/(l*(l+1)**3*(l+3)*(2*l+3)) - 3*h(l+1,m,s)**2/(l*(l+1)**3*(l+2))) if l==1 else m**3*s**6*(8*h(l,m,s)/(l**6*(l+1)**3*(l-1)**3) - 8*h(l+1,m,s)/(l**3*(l+1)**6*(l+2)**3)) + m*s**2*h(l,m,s)*(-h(l+1,m,s)*(7*l**2+7*l+4)/(l**3*(l+2)*(l+1)**3*(l-1)) - h(l-1,m,s)*(3*l-4)/(l**3*(l+1)*(2*l-1)*(l-2))) + m*s**2*((3*l+7)*(h(l+1,m,s)*h(l+2,m,s))/(l*(l+1)**3*(l+3)*(2*l+3)) - 3*h(l+1,m,s)**2/(l*(l+1)**3*(l+2)) + 3*h(l,m,s)**2/(l**3*(l-1)*(l+1)))

    return f0 + f1*c + f2*c**2 + f3*c**3 + f4*c**4 + f5*c**5
```

```python
def abg_ang(p, pars):
    
    sAlm, w, a, l, m, s = pars
    
    kp = abs(m + s)/2
    km = abs(m - s)/2
    
    c = a*w
    
    ap = -2*(p + 1)*(p + 2*km + 1) 
    bp = p*(p - 1) + 2*p*(km + kp + 1 - 2*c) - (2*c*(2*km + s + 1) - (km + kp)*(km + kp + 1)) - (c*c + s*(s + 1) + sAlm)
    gp = 2*c*(p + km + kp + s)
    
    abg = np.array([ap, bp, gp])
    
    return abg

def a_ang(p, pars):
    
    sAlm, w, a, l, m, s = pars
    
    kp = abs(m + s)/2
    km = abs(m - s)/2
    
    c = a*w
    
    return -2*(p + 1)*(p + 2*km + 1) 

def b_ang(p, pars):
    
    sAlm, w, a, l, m, s = pars
    
    kp = abs(m + s)/2
    km = abs(m - s)/2
    
    c = a*w
    
    return  p*(p - 1) + 2*p*(km + kp + 1 - 2*c) - (2*c*(2*km + s + 1) - (km + kp)*(km + kp + 1)) - (c*c + s*(s + 1) + sAlm)

def g_ang(p, pars):
    
    sAlm, w, a, l, m, s = pars
    
    kp = abs(m + s)/2
    km = abs(m - s)/2
    
    c = a*w

    return 2*c*(p + km + kp + s)


def cond_ang(sAlm, w, a, l, m, s):
    
    if not np.isscalar(sAlm): sAlm = sAlm[0]

    ct = [sAlm, w, a, l, m, s]

    f  = b_ang(0, ct)
    
    if f == 0.: f = 1.e-15
            
    C = f
    D = 0
    i = 1
    Del = 2
            
    while abs(Del-1)>1.e-20:
        
        D = b_ang(i, ct) - a_ang(i-1, ct)*g_ang(i, ct)*D
        if D == 0.: D = 1.e-15

        C = b_ang(i, ct) - a_ang(i-1, ct)*g_ang(i, ct)/C
        if C == 0.: C = 1.e-15

        D = 1./D
        Del = C*D
        f *= Del
        i += 1
    
    return f

def s_A_lm(w, a, l, m, s):
    
    #m = -m
    
    if a*w == 0.:
        
        lm = l*(l+1) - s*(s+1)
        
    elif a*w < 1.e-2:
        
        lm = lambdalms(l, m, s, a*w).llms()
        
    else:
        
        sAlm0 = lambdalms(l, m, s, a*w).llms() + 0.25*a*w*s*l

        lm = root(cond_ang, sAlm0, args=(w, a, l, m, s), method='hybr').x[0]
    
    return lm
```

### Finding Absorption Probs and $\frac{d^2 N}{d\omega dt}$ ###

```python
def alpha2(GM, a, w, m): return a**2*GM**2 + a*GM*m/w

def alpha(GM, a, w, m):

    if alpha2(GM, a, w, m) >= 0:
        return np.sqrt(alpha2(GM, a, w, m))

    else:
        return np.sqrt(-alpha2(GM, a, w, m))


def r2(r, GM, a, w, m): return r**2 + a**2*GM**2 + a*GM*m/w

def req(rs, v, GM, a, w, m):
                
    r = v[0]
                
    drdrs = (r**2 - 2*GM*r + a**2*GM**2)/(r**2 + a**2*GM**2 + a*GM*m/w)
                
    return [drdrs]

def rstar(r, GM, a, w, m):

    if a == 0.:
        return r + 2.*GM*np.log(r/(2.*GM) - 1.)
    else:
        return (r + ((2.*GM*rplus(GM,a) + a*GM*m/w)/(rplus(GM,a) - rmins(GM,a)))*np.log(r/rplus(GM,a) - 1.)
                  - ((2.*GM*rmins(GM,a) + a*GM*m/w)/(rplus(GM,a) - rmins(GM,a)))*np.log(r/rmins(GM,a) - 1.))
```

```python
def V(r, GM, a, w, sAlm, m, s):

    D = Delta(r, GM, a)
    
    rho_2 = r**2 + alpha2(GM, a, w, m)

    rho_4 = rho_2**2
    
    alp_2 = alpha2(GM, a, w, m)
    
    Vs = 0

    if abs(s) == 0.:

        Vs = (D/rho_4)*(sAlm + (D + 2.*r*(r - GM))/rho_2 - (3.*r**2*D/rho_4))

    elif abs(s) == 1.:

        if alp_2 >= 0.:
        
            Vs = (D/rho_4) * (sAlm - alp_2*D/rho_4  - 1j * sqrt(alp_2)*(2.*(r - GM)/rho_2 - 4.*r*D/rho_4))
                         
        else:
            
            alp_2 = -alp_2
            
            Vs =  (D/rho_4) * (sAlm + (D/rho_4) * alp_2 - 4*r*(D/rho_4)*sqrt(alp_2) + 2.*sqrt(alp_2)*(r - GM)/rho_2)

    return Vs
```

```python
def CDEqs(rs, v, GM, a, w, sAlm, m, s, tor_f):
    
    psi, dpsi = v
                
    ddpsi= (V(tor_f(rs), GM, a, w, sAlm, m, s) - w**2)*psi
                
    return [dpsi, ddpsi]
```

```python
def Gamma_l(w, ast, l, m, s):
    
    GM = 1.
    
    if ast >= 0.99: eps = 1.e-4
    elif GM*w > 2.e-1: eps = 1.e-4
    else: eps = 1.e-3

    inf = 1000
    
    if w > 1.: inf /= w
    
    rsminf = rstar(rplus(GM, ast) + eps, GM, ast, w, m)
    rspinf = rstar(rplus(GM, ast) + inf, GM, ast, w, m)
    
    c = ast*GM*w
    
    sAlm = lambdalms(l, -m, -s, c).llms() + c*c + 2*m*c 
    
    if w > -ast*m/(2.*rplus(GM, ast)):
        
        #print("no - superradiance")
            
        soltor = solve_ivp(lambda t, z: req(t, z, GM, ast, w, m), [rsminf, rspinf], [rplus(GM, ast) + eps], rtol=1.e-10, atol=1.e-20) 

        tor = interp1d(soltor.t, soltor.y[0], bounds_error=False, fill_value=(soltor.y[0,0], soltor.y[0,-1]))

        v0_C = [np.exp(-1j * w * rsminf), -1j * w * np.exp(-1j * w * rsminf)]

        solpsi = solve_ivp(lambda t, z: CDEqs(t, z, GM, ast, w, sAlm, m, s, tor), [rsminf, rspinf], v0_C, method='BDF', rtol=1.e-7, atol=1.e-15)
        
        psi  = solpsi.y[0,-1]
        dpsi = solpsi.y[1,-1]/(1j*w)
        
        Aout = (psi - dpsi)/(2.*np.exp(-1j*w*rspinf))
        
        Glms = 1./abs(Aout)**2
        
    else:
                
        #########################################
        #
        #      Solution in first branch
        #
        #########################################
            
        rdiv = alpha(GM, ast, w, m)
        rsdivm = rstar(rdiv - eps, GM, ast, w, m)
            
        v0_l = [rplus(GM, ast) + eps]

        soltorl = solve_ivp(lambda t, z: req(t, z, GM, ast, w, m), [rsminf, rsdivm], v0_l, method='Radau', rtol=1.e-15, atol=1.e-20)

        torl = interp1d(soltorl.t, soltorl.y[0], bounds_error=False, fill_value=(soltorl.y[0,0], soltorl.y[0,-1]))

        v0CDE_l = [np.exp(1j * w * rsminf), 1j * w * np.exp(1j * w * rsminf)]
            
        solpsil = solve_ivp(lambda t, z: CDEqs(t, z, GM, ast, w, sAlm, m, s, torl), [rsminf, rsdivm], v0CDE_l, 
                            method='BDF', rtol=1.e-7, atol=1.e-10)#, rtol=1.e-10, atol=1.e-10)
        
        #########################################
        #
        #      solution in second branch
        #
        #########################################

        rsdivp = rstar(rdiv + eps, GM, ast, w, m)
        
        v0_r = [np.sqrt(-alpha2(GM, ast, w, m)) + eps]

        soltorr = solve_ivp(lambda t, z: req(t, z, GM, ast, w, m), [rsdivp, rspinf], v0_r, method='Radau', rtol=1.e-15, atol=1.e-20)

        torr = interp1d(soltorr.t, soltorr.y[0], bounds_error=False, fill_value=(soltorr.y[0,0], soltorr.y[0,-1]))
        
        rspinf = soltorr.t[-1]

        v0CDE_r = [solpsil.y[0,-1], 0.5*solpsil.y[0,-1]*(Delta(rdiv + eps, GM, ast)/r2(rdiv + eps, GM, ast, w, m))*(1./eps)]

        solpsir = solve_ivp(lambda t, z: CDEqs(t, z, GM, ast, w, sAlm, m, s, torr), [rsdivp, rspinf], v0CDE_r, 
                            method='BDF', rtol=1.e-7, atol=1.e-10)
        
        psi = solpsir.y[0,-1]
        dpsi = solpsir.y[1,-1]/(1j*w)

        Aout = (psi - dpsi)/(2.*np.exp(-1j*w*rspinf))

        Glms = -(-1.)**(2.*s)/abs(Aout)**2

    return Glms
```

```python
def d2Ndedt(w, ast, l, m, s):
    
    Glms = Gamma_l(w, ast, l, m, s)
    
    DF = 1./(exp((w + m*Omega(GM, ast))/TBH(GM, ast)) - (-1)**(2*s))
    
    return Glms*DF/(2.*pi)
```

### Finding Angular eigenvectors --- Using Leaver's Method, following PRD 73 (2006) 024013 ###

```python
def b_k(n, pars):
    
    w, a, l, m, s = pars

    sAlm = lambdalms(l, m, -s, a*w).llms()#
    
    abg_k = array([abg_ang(i, [sAlm, w, a, l, m, s]) for i in range(3*n+1)]) 
    
    alp_k = abg_k[:,0]
    bet_k = abg_k[:,1]
    gam_k = abg_k[:,2]
    
    bk = zeros(n)
        
    bk[0] = 1.0
    bk[1] = -(bet_k[0]/alp_k[0])*bk[0]
    
    for i in range(2, n):
    
        f = 1.e-30
        
        C   = f
        D   = 0.
        Del = 0.
    
        for k in range(i, 3*n): 
        
            D = bet_k[k] - alp_k[k-1] * gam_k[k]*D
    
            if D == 0.: D += 1.e-30

            C = bet_k[k] - alp_k[k-1] * gam_k[k]/C
    
            if C == 0.: C += 1.e-30
    
            D = 1./D
    
            Del = C*D
        
            f *= Del
        
            if abs(Del - 1) < 1.e-10*bet_k[0]: break
            
        bk[i] = bk[i-1] * f/alp_k[i-1]

    mn = np.min(bk)
    mx = np.max(bk)

    if mx > abs(mn):
        bk = bk/mx
    else:
        bk = bk/mn
        
    return bk
```

```python
def Spm(th, w, a, l, m, s, n):
    
    x = cos(th)
    
    kp = abs(m + s)/2
    km = abs(m - s)/2
    
    S = 0.
    
    coeff = b_k(n, [w, a, l, m, s])
    
    for p in range(n): S += coeff[p] * (1 + x)**p
            
    S *= exp(a*w*x) * (1 + x)**km * (1 - x)**kp 
                
    return S

def Spm_n(x, w, a, l, m, s, n):
    
    kp = abs(m + s)/2
    km = abs(m - s)/2
    
    S = 0.
    
    coeff = b_k(n, [w, a, l, m, s])
    
    for p in range(n): S += coeff[p] * (1 + x)**p
            
    S *= (1 + x)**km * (1 - x)**kp  
                
    
    return 2*pi*abs(S)**2
```

### Angular distribution $\frac{d^3 N}{d\cos\theta d\omega dt}$ ###

```python
def d3Ndthdedt(w, th, ast, l, m, s):
    
    Glms =  Gamma_l(w, ast, l, m, s)
    
    n = 50
    
    Sp = Spm(th, w, ast, l, -m, -s, n)
    
    Sm = (-1)**(l-m)*Spm(th, w, ast, l, -m, -s, n)
    
    n_0 = integrate.quad(Spm_n, 0, pi, args=(w, ast, l, -m, s, n))[0]
    
    DF = 1./(exp((w + m*Omega(GM, ast))/TBH(GM, ast)) - (-1)**(2*s))
    
    return (Glms*DF/(2.*pi))*(Sp*Sp + Sm*Sm)/n_0
```
